# Day 3: Data Visualization (matplotlib and seaborn)

Let's begin with importing everything.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## 1. Matplotlib 
This library gives us 'raw control' on data. 

Different types of plots are:

### Line Plot

```python
x= [1,2,3,4,5, 6]
y= [10,20,25, 42,17,4]

plt.plot(x,y)               #x and y must have the same length
plt.title("Basic line plot")        #plot title
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)      # for grid lines

plt.savefig("../Day3-Visualization/charts/line_plot.png", dpi=300, bbox_inches='tight')
plt.show()          #for display
```

### Bar Chart

```python
names= ["Prashant", "Shaurya","Sarthak", "Lalu", "Krishna"]
ages= [21,22,22,23, 21]

plt.bar(names, ages, color="purple")
plt.title("Friends and ages bar chart")
plt.ylabel("Ages")
plt.savefig("charts/bar_chart.png", dpi=300)
plt.show()
```

### Histrogram ( Data Distribution)

```python
import numpy as np

data= np.random.randn(1000)         #graph different each time 
plt.hist(data, bins= 30, color ='skyblue')
plt.title("Histogram of random data")
plt.savefig("charts/histogram.png")
plt.show()
```

### Scatter Plot

```python
x= [5,7,2,7, 15,4,6,8,11,25]
y=[99,34,73,56,23,78,22,86,39,5]

plt.scatter(x,y,color='red')
plt.title("Scatter chart")
plt.xlabel("X here")
plt.ylabel("Y here")
plt.savefig("charts/scatter_plot.jpg")
plt.show()
```

Hence, now we have studied Line Chart, Bar Chart, Histogram and Scatter Chart- all under matplotlib. 

Let's move on to seaborn now.

## 2. Seaborn
This is built on top of matplotlib. It knows you're working with data. More advanced.

```python
tips= sns.load_dataset("tips")      # tips here is a built-in data set, no need to make a new one
tips.head()
```

### Histogram with KDE

```python
sns.histplot(tips['total_bill'], kde= True)        #KDE (Kernel Density Estimation)- the overlaying curve
plt.title("Total bill distribution")
plt.savefig("charts/histogram_sns_with_kde.jpg")
plt.show()
```

### Box Plot
used for Median, Quartiles and Outliers - the 'weird data points

```python
sns.boxplot(x='sex', y='total_bill', data=tips)
plt.title("Bill amount by sex shown")
plt.savefig("charts/box_plot_sns.png")
plt.show()
```

### Count Plot

```python
sns.countplot(x='day', data=tips)
plt.title("Total orders by day")
plt.savefig("charts/count_plot_sns.png")
plt.show()
```

### Scatter Plot (with Hue)

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips)          #hue gives different colours
plt.title("Tips vs Bill ( hue by gender )")
plt.savefig("charts/scatter_plot_sns_with_hue.jpg")
plt.show()
```

## Practice:
Let's load Day 2's data set first.

```python
df= pd.read_csv("../Day2-Pandas/cleaned_data.csv")      #use ".." to go 'up' by one level
```

Always inspect the dataframe with df.info() and df.dtypes before visualizing.

```python
#visualize correlations

#df.corr() selects only numeric data ( gives errors otherwise)
#annot=True or annotation, ensures data is displayed "inside" the cells

sns.heatmap(df[['Age', 'Score', 'Degree at age']].corr(), annot=True)
plt.savefig("charts/prac_heatmap_sns.jpg")  
plt.show()  #explicitly call plt.show() to avoid discrepancies 

sns.pairplot(df)
plt.savefig("charts/prac_pariplot_sns.jpg")
plt.show()

df.groupby('Age')['Score'].mean().plot(kind='bar')
plt.savefig("charts/prac_groupby_bar_chart.jpg")
plt.show()
```
